```python
import math


def gcd(x, y, verbose=True, steps=[]):
    if x < y:  # We want x >= y
        return gcd(y, x, verbose, steps)
    print()
    while y != 0:
        if verbose:
            data = (x, math.floor(x/y), y, x % y)
            print('%s = %s * %s + %s' % data)
            steps.append(data)
        (x, y) = (y, x % y)

    if verbose:
        print('gcd is %s' % x)
    return (x, steps)


def findOpposite(toFindOpposite, mod):
    denominator, steps = gcd(toFindOpposite, mod)

    print()
    print('==================')
    print()

    steps.pop()
    start = steps.pop()
    result = (start[3], 1, start[0], -start[1], start[2])

    print(f'{result[0]} = {result[1]} * {result[2]} + {result[3]} * {result[4]}')
    for index in range(len(steps) - 1, -1, -1):
        start = (steps[index][3], steps[index][0], -steps[index][1], steps[index][2])
  
        rem_1, res_coef_1, res_1, koef_x_1, x_1 = result
        rem_2, res_2, koef_x_2, x_2 = start

        result = (
            rem_1,
            koef_x_1,
            res_2,
            res_coef_1 + (koef_x_1 * koef_x_2),
            res_1
        )

        print(f'{result[0]} = {result[1]} * {result[2]} + {result[3]} * {result[4]}')

    print('soulution: ', result[3])
    print('abs: ', result[3] if result[3] > 0 else result[3] + mod)

findOpposite(185, 311)
```

```python
getBin = lambda x: format(x, 'b')


def findPowerByMod(x, power, mod):
    binaryPower = getBin(power)

    reversedBinaries = [(index, binary) for index, binary in enumerate(binaryPower[::-1])][::-1]
    print(f"{power} = ", [f"{binary} * 2^{index}" for index, binary in reversedBinaries])

    reversedBinariesMinus1 = reversedBinaries[1::]
    res = x
    for (index, binary) in reversedBinariesMinus1:
        print()
        print(f"i = {index}, binary: {binary}")

        total = pow(res, 2)
        totalByMod = total % mod
        print(f"{res}^2 mod {mod} = {totalByMod}")

        if binary == '0':
            res = totalByMod
        else:
            print(f"Since BiT == 1: ({totalByMod} * {x}) % {mod} = {(totalByMod * x) % mod}")
            res = (totalByMod * x) % mod

    print(f"result = {res}")

x = 11
power = 83
mod = 105
findPowerByMod(x, power, mod)
```

```python
# Символ Лежандра

def isPrime(n):
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
        
    return True

def isSquare(a):
    if a < 0:
        return False
    return a ** .5 == int(a**0.5)


def is2SquareByMod(mod):
    temp = mod % 8
    tempArr = [temp, -temp]
    print(f"{mod} = 8 * n + {temp}")

    if -1 in tempArr or 1 in tempArr:
        print(f"{mod} ^ 2 / 8 - is even")
        return 1

    if -3 in tempArr or 3 in tempArr:
        print(f"{mod} ^ 2 / 8 - is odd")
        return -1


def isSquareByMod(a, mod):
    if not isPrime(a):
        print(f"Numerator {a} is not prime")

    counter = 0
    result = None
    hasMinus1 = False

    while result not in [-1, 1]:
        if a < 0:
            hasMinus1 = not hasMinus1
            a = -a

        print(f"-1 * {a} / {mod}") if hasMinus1 else print(f"{a} / {mod}")

        if isSquare(a):
            result = 1 * (-1 if hasMinus1 else 1)
            break

        isModPrime = isPrime(mod)
        isAEven = a % 2 == 0
        if isModPrime and not isAEven:
            minusOnePower = (a - 1) * (mod -1) / 4

            a, mod = mod, a
            print(f"-1^(({a} - 1) * ({mod} -1) / 4) * {a} / {mod}")

            a = a % mod
            if minusOnePower % 2 == 1:
                a = -a
                print(f"{a} / {mod}")
            
            continue

        if isAEven:
            print()
            print(f'----- (2 / {mod})')
            a = a // 2

            print(f"(2 / {mod}) * {a} / {mod}")

            is2Square = is2SquareByMod(mod)
            print(f"is 2 square: {is2Square}")

            a = a * is2Square
            print(f'/----- (2 / {mod})')
            print()

            print(f"{a} / {mod}")

            continue

        if not isModPrime:
            print(f"module: {mod} is not prime. Inversing numerator")
            a = a - mod

        # If cannot solve, stop
        counter = counter + 1
        if (counter > 25):
            break

    print(f"result = {result}")


isSquareByMod(358, 571)
```

```python
# SYSTEM OF EQUATIONS 

def findInverse(n, mod):
    print(f"Find {n}^-1 mod {mod}")
    print(f"{n} * x = {mod}k + 1")
    
    koef = 1
    while True:
        res = ((mod * koef) + 1) / n
        print(f"Trying {mod} * {koef} + 1 = {(mod * koef) + 1}")

        if (res == int(res)):
            print(f"Found: {mod} * {koef} + 1 = {(mod * koef) + 1}") 
            return int(res)
        else:
            koef = koef + 1


def getValidTable(table):
    validTable = []

    for (a, res, mod) in table:
        if a != 1:
            inverse = findInverse(a, mod)
            print(f"inverse = {inverse}")

            a = 1
            res = (res * inverse)
            res = res % mod if res > 0 else (res % mod) - mod

            print(f"{a}x = {res} (mod {mod})")
            print()

        validTable.append([a, res, mod])

    print()
    print()
    for (a, res, mod) in validTable:
        print(f"{a}x = {res} (mod {mod})")

    return validTable


def greatChineeseTheorem(table):
    for (a, res, mod) in table:
        print(f"{a}x = {res} (mod {mod})")

    print()
    print()


    validTable = getValidTable(table) if len([a for (a, res, mod) in table if a != 1]) > 0 else table

    M = 1
    for (_, _, mod) in validTable:
        M = M * mod

    MxList = []
    for (_, _, mod) in validTable:
        MxList.append(M // mod)

    print(f"M = {[mod for (_, _, mod) in validTable]} = {M}")
    print()

    print(f"MxList = {MxList}")

    print()
    print()

    CList = []
    for index, Mx in enumerate(MxList):
        (a, res, mod) = validTable[index]

        MxMod = Mx % mod if Mx > 0 else (Mx % mod) - mod
        print(f'C{index} = {Mx}^-1 mod({mod}) = {MxMod}^-1 mod({mod})')
        
        if MxMod in [-1, 1]:
            CList.append(MxMod)
        else:
            inverse = findInverse(MxMod, mod)
            CList.append(inverse)
        print('---')

    print(f"CList = {CList}")
    print()
    print()

    result = 0
    print(f"X = ")
    for index, Mx in enumerate(MxList):
        (a, res, mod) = validTable[index]
        C = CList[index]

        result += res * Mx * C
        print(f"+ {res} * {Mx} * {C}")
        


    print(f"= {result} (mod {M})")
    print(f"= {result % M} (mod {M})")


table = [
    [1, 0, 2],
    [1, 3, 7],
    [1, 5, 9],
    [1, -1, 5],
]
greatChineeseTheorem(table)
    
```
